# Laplacian Distribution Per Image

## 0. Import Libraries

```python
import os
import cv2
import matplotlib.pyplot as plt
import seaborn as sns
from video_to_frame import *
```

## 1. Read All Extracted Frames, Plot Distribution
Prior to the execution, the `video_to_frames(video_path, save_path)` should be executed first in `video_to_frame.py`.

```python
frame_vars = []
img_path = './box_frames/all_frames/'

for file in sorted(os.listdir(img_path)):
    image = cv2.imread(img_path + file)
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    laplacian_var = variance_of_laplacian(gray)
    frame_vars.append(laplacian_var)

print('Min:', min(frame_vars))
print('Max:', max(frame_vars))
```

```python
sns.distplot(frame_vars, kde=True, hist=True)
plt.title("Histogram of Laplacian Variance Distribution")
plt.xlabel("Variance")
plt.ylabel("Count")
plt.show()
```

## 2. Top 40% of Laplacian Variance
The `min` value that it outputs will be used as a `threshold` value to cut off and select "good" frames from all extracted frames.

```python
frame_vars.sort(reverse=True) # sort in descending order
index_top_40_per = int(len(frame_vars) * 0.04)
top_40_var = frame_vars[:index_top_40_per]

sns.distplot(top_40_var, kde=True, hist=True)
plt.title("Histogram of Top 20% Laplacian Variance Distribution")
plt.xlabel("Variance")
plt.ylabel("Count")
plt.show()

print('Min:', min(top_40_var))
print('Max:', max(top_40_var))
```

## 3. Linspace first (Evenly selected -> 5%)

```python
frame_vars = []
img_path = './box_frames/linspace/'

for file in sorted(os.listdir(img_path)):
    image = cv2.imread(img_path + file)
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    laplacian_var = variance_of_laplacian(gray)
    frame_vars.append(laplacian_var)

print('Min:', min(frame_vars))
print('Max:', max(frame_vars))

sns.distplot(frame_vars, kde=True, hist=True)
plt.title("Histogram of Laplacian Variance Distribution")
plt.xlabel("Variance")
plt.ylabel("Count")
plt.show()
```

## 4. Select 30% of Linspace

```python
frame_vars.sort(reverse=True) # sort in descending order
index_top_30_per = int(len(frame_vars) * 0.3)
top_30_var = frame_vars[:index_top_30_per]

sns.distplot(top_30_var, kde=True, hist=True)
plt.title("Histogram of Top 20% Laplacian Variance Distribution")
plt.xlabel("Variance")
plt.ylabel("Count")
plt.show()

print('Min:', min(top_30_var))
print('Max:', max(top_30_var))
```
